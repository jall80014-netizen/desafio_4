```python
import json
import random

# ------------------ Utilidades ------------------

def tirar_dado9():
    return random.randint(1, 9)

def evaluar_resultado(valor):
    if 1 <= valor <= 3:
        return "Falla total"
    elif 4 <= valor <= 5:
        return "Casi fallo"
    elif 6 <= valor <= 7:
        return "Casi sale bien"
    elif 8 <= valor <= 9:
        return "¡Perfecto!"

def cargar_personajes():
    try:
        with open("personajes.json", "r", encoding="utf-8") as f:
            return json.load(f)
    except FileNotFoundError:
        print("Error: 'personajes.json' not found. Creating an empty list.")
        return []

def guardar_personajes(personajes):
    with open("personajes.json", "w", encoding="utf-8") as f:
        json.dump(personajes, f, indent=4, ensure_ascii=False)

def cargar_acciones():
    try:
        with open("acciones.json", "r", encoding="utf-8") as f:
            return json.load(f)
    except FileNotFoundError:
        print("Error: 'acciones.json' not found. Creating an empty list.")
        return []

def mostrar_tabla(personajes):
    print("\n📊 Tabla de Clasificación:")
    if not personajes:
        print("No hay jugadores registrados todavía.")
        return
    ordenados = sorted(personajes, key=lambda x: x["puntaje"], reverse=True)
    for i, p in enumerate(ordenados, 1):
        print(f"{i}. {p['usuario']} ({p['personaje']}) - Puntaje: {p['puntaje']}")

# ------------------ Eventos ------------------

def evento_dueno(personaje):
    lugares = ["cama", "cocina", "sala"]
    lugar_dueno = random.choice(lugares)
    print(f"\n🚨 ¡El dueño se acerca! Está en la {lugar_dueno}.")
    escondite = input("¿Dónde te escondes? (cama/cocina/sala): ").lower()

    if escondite not in lugares:
        print("❌ No elegiste bien, el dueño te atrapó.")
        personaje["puntaje"] -= 10
        return True

    prob = 0.8 if escondite == lugar_dueno else 0.2
    if random.random() < prob:
        print("💀 El dueño te encontró... ¡Fin del juego!")
        personaje["puntaje"] -= 5
        return True
    else:
        print("😺 Lograste esconderte con éxito.")
        personaje["puntaje"] += 5
        return False

def piedra_papel_tijera(personaje):
    opciones = ["piedra", "papel", "tijera"]
    jugador = input("Elige piedra, papel o tijera: ").lower()
    if jugador not in opciones:
        print("❌ Opción inválida.")
        return False

    pescado = random.choice(opciones)
    print(f"🐟 El pescado eligió: {pescado}")

    if jugador == pescado:
        print("🤝 Empate, el pescado se escapó.")
        personaje["puntaje"] -= 1
        return False
    elif (jugador == "piedra" and pescado == "tijera") or \
         (jugador == "papel" and pescado == "piedra") or \
         (jugador == "tijera" and pescado == "papel"):
        print("🎉 ¡Ganaste! Atrapas el pescado.")
        personaje["puntaje"] += 10
        personaje["inventario"].append("Pescado atrapado")
        return True
    else:
        print("💀 Perdiste, el pescado escapó.")
        personaje["puntaje"] -= 5
        personaje["energia"] -= 2
        return False

def vigilar_pescado(personaje):
    print("\n🐱 Otro gato se acerca a tu pescado...")
    decision = input("¿Quieres vigilarlo? (si/no): ").lower()
    prob = 0.2 if decision == "si" else 0.6

    if random.random() < prob:
        print("💀 El otro gato te robó el pescado.")
        if "Pescado atrapado" in personaje["inventario"]:
            personaje["inventario"].remove("Pescado atrapado")
        personaje["puntaje"] -= 10
        return False
    else:
        print("😺 Lograste proteger tu pescado.")
        personaje["puntaje"] += 5
        return True

def defender_pescado(personaje):
    print("\n🐱 Otros gatos se acercan demasiado...")
    decision = input("¿Quieres pegarles para que se vayan? (si/no): ").lower()
    prob = 0.1 if decision == "si" else 0.5
    if decision == "si":
        personaje["energia"] -= 1
        print("👊 Les pegaste y gastaste energía.")

    if random.random() < prob:
        print("💀 Los gatos te robaron el pescado.")
        if "Pescado atrapado" in personaje["inventario"]:
            personaje["inventario"].remove("Pescado atrapado")
        personaje["puntaje"] -= 10
        return False
    else:
        print("😺 Mantienes tu pescado seguro.")
        personaje["puntaje"] += 5
        return True

def entregar_pescado(personaje):
    print("\n🐱 Vas a buscar a la Gata Blanca...")
    if "Pescado atrapado" in personaje["inventario"]:
        print("🎉 La Gata Blanca acepta tu pescado y ahora es tu pareja.")
        personaje["puntaje"] += 20
        personaje["suerte"] += 2
        personaje["inventario"].remove("Pescado atrapado")
        personaje["inventario"].append("Pareja: Gata Blanca")
        return True
    else:
        print("💔 No tienes pescado... La Gata Blanca se enoja.")
        personaje["puntaje"] -= 10
        return False

# ------------------ Niveles ------------------

def nivel1(personaje):
    print("\n🐾 Nivel 1: Conseguir pescado y robar leche")
    objetivos = ["Agarrar un pescado", "Robar la leche"]
    completados = []

    while len(completados) < len(objetivos) and personaje["energia"] > 0:
        accion = input("Elige acción (pescado/leche): ").lower()
        dado = tirar_dado9()
        resultado = evaluar_resultado(dado)
        print(f"Tiraste el dado... salió {dado} 🎲 → {resultado}")

        if resultado == "Falla total":
            personaje["puntaje"] -= 5
            personaje["energia"] -= 2
        elif resultado == "Casi fallo":
            personaje["puntaje"] -= 2
            personaje["energia"] -= 1
        elif resultado == "Casi sale bien":
            personaje["puntaje"] += 5
            personaje["energia"] -= 1
            if accion == "pescado":
                completados.append("Agarrar un pescado")
            elif accion == "leche":
                completados.append("Robar la leche")
        elif resultado == "¡Perfecto!":
            personaje["puntaje"] += 10
            personaje["suerte"] += 1
            personaje["energia"] -= 1
            if accion == "pescado":
                completados.append("Agarrar un pescado")
            elif accion == "leche":
                completados.append("Robar la leche")

        print(f"⚡ Energía actual: {personaje['energia']}")

    return len(completados) == len(objetivos)

def nivel2(personaje):
    print("\n🐾 Nivel 2: Atrapar pescado, vigilarlo, defenderlo y entregarlo a la Gata Blanca")
    completados = []

    atrapado = piedra_papel_tijera(personaje)
    if atrapado:
        vigilar_pescado(personaje)
        defender_pescado(personaje)
        entregar_pescado(personaje)
        completados.append("Atrapar y entregar pescado a Gata Blanca")

    return len(completados) > 0

# ------------------ Juego principal ------------------

def jugar():
    personajes = cargar_personajes()
    usuario = input("Ingresa tu nombre de usuario: ")

    personaje = next((p for p in personajes if p["usuario"].lower() == usuario.lower()), None)
    if personaje is None:
        nombre_personaje = input("Elige tu personaje (ej: Gato Gris): ")
        personaje = {
            "usuario": usuario,
            "personaje": nombre_personaje,
            "puntaje": 0,
            "energia": 10,
            "hambre": 5,
            "suerte": 3,
            "fuerza": 4,
            "inventario": []
        }
        personajes.append(personaje)

    if nivel1(personaje):
        print("\n🎉 ¡Has completado el Nivel 1! Pasas al Nivel 2.")
        nivel2(personaje)
    else:
        print("\n💀 No lograste completar el Nivel 1.")

    guardar_personajes(personajes)
    mostrar_tabla(personajes)

# ------------------ Menú ------------------

def menu():
    while True:
        print("\n🐱 Menú del Juego de Gato")
        print("1. Jugar")
        print("2. Ver clasificación")
        print("3. Salir")
        opcion = input("Elige una opción (1-3): ")

        if opcion == '1':
            jugar()
        elif opcion == '2':
            personajes = cargar_personajes() # Load characters to show current ranking
            mostrar_tabla(personajes)
        elif opcion == '3':
            print("¡Gracias por jugar!")
            break
        else:
            print("Opción inválida. Por favor, elige un número del 1 al 3.")
```
